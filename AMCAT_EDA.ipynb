```python
# Step 1: Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_numerical['Salary'].quantile(0.25)
Q3 = df_numerical['Salary'].quantile(0.75)
IQR = Q3 - Q1  # Interquartile range

# Step 2: Define the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Identify outliers
outliers = df_numerical[(df_numerical['Salary'] < lower_bound) | (df_numerical['Salary'] > upper_bound)]

# Step 4: Remove outliers
df_cleaned = df_numerical[(df_numerical['Salary'] >= lower_bound) & (df_numerical['Salary'] <= upper_bound)]
df_cleaned.head()
```
